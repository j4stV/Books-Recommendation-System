```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/preprocessed_data.csv')
data.head(5)
```

```python
books_data = data[['ISBN', 'raw_title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'categories']]
books_data = books_data.drop_duplicates()
books_data.head(5)
```

```python
user_data = data[['User-ID', 'Location', 'Age']]
user_data = user_data.drop_duplicates()
user_data.head(5)
```

```python
books_data.to_csv("data/books_data.csv", index=False)
user_data.to_csv("data/user_data.csv", index=False)
```
