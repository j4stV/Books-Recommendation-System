```python
import pandas as pd
import numpy as np
import ast
from warnings import simplefilter
```

```python
simplefilter('ignore')
```

```python
data = pd.read_csv('data/data.csv')
data.head(5)
```

```python
data = data.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L', 'description', 'raw_year', 'Unnamed: 0', 'Book-Title'], axis=1)
data.head(5)
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['Publisher'])
```

```python
data['categories'] = data['categories'].apply(ast.literal_eval)
data['categories'] = data['categories'].apply(lambda x: sorted(x))
data['categories'] = data['categories'].apply(lambda x: ''.join(x))
data.head(5)
```

```python
data['Location'] = data['Location'].apply(lambda x: x.split(',')[-1])
data.head(5)
```

```python
def clearLocation(a):
    if len(a) > 0 and a[0] == ' ':
        a = a[1:]
    unknown = ['', 'n/a', 'u.s>', 'k1c7b1', 'la chine éternelle!', 'the world tomorrow', 'here and there', 'the', 'everywhere and anywhere', '"n/a\\"', 'quit', 'usa (currently living in england)', 'far away...', 'n/a - on the road', 'ee.uu', 'we`re global!', 'c', 'x', 'c.a.', 'orange co', 'space', '"', 'lkjlj', 'universe']
    usa = ['u.s.a.', 'ysa', 'united staes', 'san franicsco', 'united states of america', 'united states', 'u.s.a>', 'usa & canada', 'united sates', 'america', 'united state', 'us', 'hungary and usa', 'u.s. of a.']
    nz = ['nz']
    uk = ['u.k.']
    uae = ['u.a.e']
    if a in unknown:
        return 'unknown'
    elif a in usa:
        return 'usa'
    elif a in nz:
        return 'new zealand'
    elif a in uk:
        return 'united kingdom'
    elif a in uae:
        return 'united arab emirates'
    elif a[-1] == '"':
        return a[:-1]
    else:
        return a
```

```python
data['Location'] = data['Location'].apply(clearLocation)
data.head(5)
```

```python
data = data[data['Book-Rating'] != 0]
data.head(5)
```

```python
data['Publisher'] = data['Publisher'].apply(lambda x: x.replace(' ', '').lower())
data['Book-Author'] = data['Book-Author'].apply(lambda x: x.replace(' ', '').lower())
data['categories'] = data['categories'].apply(lambda x: x.replace(' ', '').lower())
data.head(5)
```

```python
data[data['Year-Of-Publication'] == 0]
```

```python
data['Year-Of-Publication'] = np.where(data['Year-Of-Publication'] == 0, np.nan, data['Year-Of-Publication'])
data['Year-Of-Publication'] = data.groupby('Publisher')['Year-Of-Publication'].transform(lambda x: x.fillna(x.median()))
data.head(5)
```

```python
data = data.drop_duplicates(subset=['User-ID', 'ISBN'])
data.head(5)
```

```python
data.to_csv('data/preprocessed_data.csv', index=False)
```
